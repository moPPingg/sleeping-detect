```python
# Import thÆ° viá»‡n
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cáº¥u hÃ¬nh matplotlib Ä‘á»ƒ hiá»ƒn thá»‹ trong notebook
%matplotlib inline
plt.style.use('seaborn-v0_8-darkgrid')

print("âœ… ÄÃ£ import táº¥t cáº£ thÆ° viá»‡n!")

```

```python
# Load dataset
DATASET_FILE = "dataset_full.csv"

try:
    data = pd.read_csv(DATASET_FILE)
    print(f"âœ… ÄÃ£ load dataset: {len(data)} dÃ²ng dá»¯ liá»‡u")
except FileNotFoundError:
    print(f"âŒ Lá»–I: KhÃ´ng tÃ¬m tháº¥y file '{DATASET_FILE}'")
    print("ðŸ’¡ HÃ£y cháº¡y notebook 01_Data_Collection.ipynb trÆ°á»›c Ä‘á»ƒ táº¡o dataset")

```

## 1. Thá»‘ng kÃª cÆ¡ báº£n



```python
# 1. Thá»‘ng kÃª cÆ¡ báº£n
print("\n1. THá»NG KÃŠ CÆ  Báº¢N:")
print(data.describe())

```

## 2. PhÃ¢n bá»‘ Label



```python
# 2. PhÃ¢n bá»‘ label
print("\n2. PHÃ‚N Bá» LABEL:")
label_counts = data['Label'].value_counts()
print(label_counts)
print(f"Tá»· lá»‡: {data['Label'].value_counts(normalize=True)}")

if len(label_counts) == 2:
    imbalance = abs(label_counts[0] - label_counts[1]) / len(data)
    if imbalance > 0.2:
        print(f"âš ï¸ Dá»¯ liá»‡u khÃ´ng cÃ¢n báº±ng! (ChÃªnh lá»‡ch: {imbalance*100:.1f}%)")
    else:
        print("âœ… Dá»¯ liá»‡u khÃ¡ cÃ¢n báº±ng")

# Váº½ biá»ƒu Ä‘á»“ phÃ¢n bá»‘
plt.figure(figsize=(8, 5))
label_counts.plot(kind='bar', color=['green', 'red'])
plt.title('PhÃ¢n bá»‘ Label')
plt.xlabel('Label (0=Tá»‰nh tÃ¡o, 1=Buá»“n ngá»§)')
plt.ylabel('Sá»‘ lÆ°á»£ng')
plt.xticks(rotation=0)
for i, v in enumerate(label_counts):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.tight_layout()
plt.show()

```

## 3. Kiá»ƒm tra Missing Values vÃ  Outliers



```python
# 3. Kiá»ƒm tra missing values
print("\n3. KIá»‚M TRA MISSING VALUES:")
missing = data.isnull().sum()
if missing.sum() > 0:
    print(missing[missing > 0])
else:
    print("âœ… KhÃ´ng cÃ³ dá»¯ liá»‡u bá»‹ thiáº¿u")

# 4. Kiá»ƒm tra outliers
print("\n4. KIá»‚M TRA OUTLIERS:")
features = ['EAR', 'MAR', 'Pitch', 'Yaw', 'Roll']
for col in features:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((data[col] < lower_bound) | (data[col] > upper_bound)).sum()
    print(f"  {col}: {outliers} outliers ({outliers/len(data)*100:.1f}%)")

```

## 4. Correlation Matrix



```python
# 5. Correlation matrix
print("\n5. CORRELATION MATRIX:")
corr = data.corr()
print(corr)

# Váº½ heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

```

## 5. Visualization - PhÃ¢n bá»‘ theo Label



```python
# Váº½ boxplot cho tá»«ng feature theo label
features = ['EAR', 'MAR', 'Pitch', 'Yaw', 'Roll']
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, feature in enumerate(features):
    ax = axes[idx]
    data.boxplot(column=feature, by='Label', ax=ax)
    ax.set_title(f'{feature} by Label')
    ax.set_xlabel('Label (0=Tá»‰nh, 1=Ngá»§)')
    ax.set_ylabel(feature)

# XÃ³a subplot thá»«a
fig.delaxes(axes[5])

plt.suptitle('PhÃ¢n bá»‘ Features theo Label', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

```
